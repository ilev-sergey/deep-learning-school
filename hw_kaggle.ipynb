# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров другого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
X = data.drop(columns="Churn")
y = data["Churn"]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 42)
```

```python
# One Hot Encoding
categorical_transformer = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

one_hot_encoded = categorical_transformer.fit_transform(X_train[cat_cols])

one_hot_df = pd.DataFrame(one_hot_encoded, columns=categorical_transformer.get_feature_names_out(cat_cols), index=X_train.index)

X_train_encoded = pd.concat([X_train, one_hot_df], axis=1)
X_train_encoded = X_train_encoded.drop(cat_cols, axis=1)

# OHE for X_test
one_hot_encoded = categorical_transformer.transform(X_test[cat_cols])

one_hot_df = pd.DataFrame(one_hot_encoded, columns=categorical_transformer.get_feature_names_out(cat_cols), index=X_test.index)

X_test_encoded = pd.concat([X_test, one_hot_df], axis=1)
X_test_encoded = X_test_encoded.drop(cat_cols, axis=1)
```

```python
# Feature Scaling
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_encoded)
X_test_scaled = scaler.transform(X_test_encoded)
```

```python
# Hyperparameter selection
X_train_split, X_validation, y_train_split, y_validation = train_test_split(X_train_scaled, y_train, train_size = 0.8, random_state = 42)

CV_model = LogisticRegressionCV(scoring="roc_auc", refit=True, n_jobs=-1)
CV_model.fit(X_train_split, y_train_split)

CV_model.scores_
```

Выпишите какое лучшее качество и с какими параметрами вам удалось получить

```python
CV_model.score(X=X_validation, y=y_validation), CV_model.score(X=X_train_split, y=y_train_split)
```

```python
train_auc = roc_auc_score(y_train, CV_model.predict(X_train_scaled))
test_auc = roc_auc_score(y_test, CV_model.predict(X_test_scaled))

test_auc, train_auc
```
